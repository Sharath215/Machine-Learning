```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("insurance_pre.csv")
dataset 
```

```python
dataset = pd.get_dummies(dataset, dtype=int)
```

```python
dataset
```

```python
dataset.columns
```

```python
independent = dataset[['age', 'bmi', 'children','sex_female', 'sex_male','smoker_no', 'smoker_yes']]
independent
```

```python
dependent = dataset[['charges']]
```

```python
dependent
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(independent,dependent,test_size = 1/3, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVR  # importing Support vector machine model.

param_grid = {'kernel':['rbf','poly','sigmoid','linear'],'C':[10,100,1000,2000,3000],'gamma':['auto','scale']}

grid = GridSearchCV(SVR(),param_grid, refit = True, verbose = 3, n_jobs=-1)


grid.fit(X_train,y_train)
```

```python
# grid_predictions = grid.predict(X_test)
# from sklearn.metrics import r2_score
# r_score = r2_score(y_test, grid_predictions)

re = grid.cv_results_

print("the R score value for best parameter {}:".format(grid.best_params_))
```
