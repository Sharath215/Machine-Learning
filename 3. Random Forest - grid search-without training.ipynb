```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("insurance_pre.csv")
dataset 
```

```python
dataset = pd.get_dummies(dataset, dtype=int)
```

```python
dataset
```

```python
dataset.columns
```

```python
independent = dataset[['age', 'bmi', 'children','sex_female', 'sex_male','smoker_no', 'smoker_yes']]
independent
```

```python
dependent = dataset[['charges']]
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor  # importing Support vector machine model.

param_grid = {'criterion':['absolute_error', 'squared_error', 'poisson', 'friedman_mse'],'max_features':['auto','sqrt','log2'],'n_estimators':[10,100]}

grid = GridSearchCV(RandomForestRegressor(),param_grid, refit = True, verbose = 3, n_jobs=-1)


grid.fit(independent,dependent)
```

```python
# grid_predictions = grid.predict(X_test)
# from sklearn.metrics import r2_score
# r_score = r2_score(y_test, grid_predictions)

re = grid.cv_results_

print("the R score value for best parameter {}:".format(grid.best_params_))
```

```python
table = pd.DataFrame.from_dict(re)
table
```

```python
Future_Prediction = grid.predict([[age_input,bmi_input,children_input,sex_male_input,smoker_yes_input,sex_Female_input,smoker_No_input]])
print("Future_Prediction={}".format(Future_Prediction))
```
